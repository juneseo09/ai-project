🪻🪻 데이터 소개(붓꽃-iris)
품종 3가지: setosa(세토사) versicolor, virginica

각 꽂은 다음 4가지 특성
1.꽃 받침 길이(sepal length)
2.꽃받침 너비(sepal width)
3.꽃잎 길이(petal length)
4.꽃잎 너비(petal width)

```python
#step1. 데이터 불러오기
iris = load_iris()
X = iris.data
Y = iris.target
#print(X)
#print(Y)

#데이터 display
pd.DataFrame(X, columns = iris.feature_names).head()
```

#ai가 공부할 데이터 준비
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

```python
#ai가 공부할 데이터 준비
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)
```

```python
#step3: Ai 모델 만들기
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
#step4: Ai가 test를 보게하기
Y_pred = model.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)
print('Accuracy:', acc*100, '&')
```

```python
#Step5 직접 꽃 크기 입력(sepal, petal)
#sample data
#5.1 3.5 1.4 0.2 setosa
#6.0 2.2 4.0 1.0 versicolor
#7.7 3.0 6.1 2.3 versinica
print('꽃의 크기를 입력해주서요!')
sl = float(input('꽃받침 길이(sepal length):'))
sw = float(input('꽃받침 너비(sepal width):'))
pl = float(input('꽃잎 길이(petal length):'))
pw = float(input('꽃받침 길이(petal width):'))
user_input = [[sl , sw , pl , pw]]
prediction = model.predict(user_input)
print('prediction data value:', prediction)
print("네, 예측된 품종은:", iris.target_names[prediction[0]])
```
