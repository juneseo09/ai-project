그룹별 분석

```python
#groupby()
#성별에따른 생존률의 평균
df.groupby('Sex')['Survived'].mean()
```

```python
#선실 등급(Pclass)에 따른 평균 나이
df.groupby('Pclass')['Age'].mean()
```

#간단한 시각화

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar' . title='Survival rate by Sex')
```
